```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins_data= pd.read_csv("pengu.csv")
penguins_data =penguins_data.dropna()
```

```python
penguins_data.shape
print(penguins_data.head())

print(penguins_data.shape)

print(penguins_data["Culmen Depth (mm)"])


```

```python
plt.hist(penguins_data["Culmen Depth (mm)"],color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
flipper_length_data = penguins_data["Flipper Length (mm)"].dropna()

# Create a figure for the box plot
plt.figure(figsize=(6, 6))

# Generate the box plot for Flipper Length
plt.boxplot(flipper_length_data, patch_artist=True)
```

```python
flipper_length_data = penguins_data["Flipper Length (mm)"].dropna()


# Create a figure for the violin plot

# Generate the violin plot for Flipper Length
sns.violinplot(x=flipper_length_data, color='lightblue')

# Add title and labels
plt.title("Violin Plot of Flipper Length (mm)", fontsize=16)
plt.ylabel("Flipper Length (mm)", fontsize=12)

# Show the plot
plt.show()
```

```python
sns.countplot(x='Island', data=penguins_data, palette='Set2')

# Add title and labels
plt.title("Bar Plot of Penguin Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Show the plot
plt.show()
```
